```python

import csv



with open('./js/bun/bun-stats-r3.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    with open('./js/bun/bun-stats-conv-r3.csv', 'w', newline='') as csvfile2:
        fieldnames = ['name', 'cpu', 'memory']
        writer = csv.DictWriter(csvfile2, fieldnames=fieldnames)

        writer.writeheader()

        for row in reader:
            print(row)

            mem1 = row[' memory'].strip()
            strp_index = 0
            for i in range(len(mem1)):
                if mem1[i] == "M":
                    strp_index = i
                    break
            mem_usage = float(mem1[0:i])

            writer.writerow({'name': row['name'], 'cpu': float(row[' cpu'].strip('%')) / 100, 'memory': mem_usage})
```
